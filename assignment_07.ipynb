## load point data for training and testing

```python
fname_data  = 'assignment_07_data.csv'

data        = np.genfromtxt(fname_data, delimiter=',')
num_data    = data.shape[0]

xx          = np.zeros(num_data)
yy          = np.zeros(num_data)
zz          = np.zeros(num_data)

for i in range(num_data):

    xx[i] = data[i,0]
    yy[i] = data[i,1]
    zz[i] = data[i,2]
    
print(data)
print(num_data)
print(xx.shape[0])
print(np.transpose(xx))
print(yy)
print(zz)
```

## compute the loss function

```python
def compute_residual(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    
    f = theta[0] + theta[1]*x + theta[2]*y
    residual = z - f
    num_data = x.shape[0]


    # ============================
    return residual, num_data
```

```python
def compute_loss(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    
    residual, n = compute_residual(theta, x, y, z)
    
    
    loss = (1/(2*n))*np.sum(residual**2)
    # ============================
    return loss
```

## compute the gradient for each model parameter (DO NOT COMPUTE THE GRADIENT FOR EACH MODEL PARAMETER, BUT DO COMPUTE THE GRADIENT OF THE MODEL PARAMTER VECTOR)

```python
def compute_gradient(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    
    
    one = np.ones((num_data,1))
    x = np.reshape(x,(num_data,1))
    y = np.reshape(y,(num_data,1))
    
    
    X = np.hstack((one,x))
    X = np.hstack((X,y))
    #print("theta[0]",theta[0])
    #print("X.shpae", X.shape)
    #print("np.dot(X,theta).shape",np.dot(X,theta).shape)
    #print("(np.dot(X,theta) - z).shape", (np.dot(X,theta) - z).shape)
    #print("grad",(np.dot( np.transpose(X), np.dot(X,theta) - z )).shape )
    grad = (2/num_data)*np.dot( np.transpose(X), np.dot(X,theta) - z ) 
    #print(grad.shape)
    # ============================
    return grad
```

## gradient descent for each model parameter

```python
num_iteration   = 1000
learning_rate   = 0.01

theta           = np.array((0, 0, 0))
theta_iteration = np.zeros((num_iteration, theta.size))
loss_iteration  = np.zeros(num_iteration)
#print(theta)
#print(theta.size)
#print(theta.shape)
for i in range(num_iteration):
    # ===== FILL UP THE CODE =====
    
    
    theta = theta - learning_rate * compute_gradient(theta,xx,yy,zz)
    loss = compute_loss(theta, xx, yy, zz)
    
    
    theta_iteration[i] = theta
    loss_iteration[i] = loss
    
    # ============================
    print("iteration = %4d, loss = %5.5f" % (i, loss))
    
```

## plot the results

```python
def plot_loss_curve(loss_iteration):

    plt.figure(figsize=(8,6))
    plt.title('loss')
    # ===== FILL UP THE CODE =====

    plt.plot(loss_iteration, color = 'red')

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
def plot_data(xx, yy, zz):
    
    fig = plt.figure(figsize=(12, 8))
    ax  = plt.subplot(111, projection='3d')
    plt.title('data')
    # ===== FILL UP THE CODE =====

    plt.plot(xx,yy,zz,'.',color = 'blue')
    # ============================    
    plt.tight_layout()
    plt.show()
```

```python

def plot_model_parameter(theta_iteration):

    plt.figure(figsize=(8,6))
    plt.title('model parameter')
    # ===== FILL UP THE CODE =====
    #theta_iteration
    plt.plot(theta_iteration[:,0],color = 'red')
    plt.plot(theta_iteration[:,1],color = 'green')
    plt.plot(theta_iteration[:,2],color = 'blue')

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
plot_loss_curve(loss_iteration)
```

```python
plot_model_parameter(theta_iteration)
```

```python
# ===== FILL UP THE CODE =====

X0 = np.arange(-15, 15, 0.1)
X1 = np.arange(-15, 15, 0.1)
XX, YY = np.meshgrid(X0,X1)
ZZ = theta[0] + theta[1]*XX + theta[2] *YY
# ============================

def plot_surface(XX, YY, ZZ, xx, yy, zz):
    fig = plt.figure(figsize=(8,8))
    ax  = fig.add_subplot(111, projection='3d')
    plt.title('regression surface')
    ax  = plt.axes(projection='3d')
    
    # ===== FILL UP THE CODE =====
    ax.plot_surface(XX,YY,ZZ,alpha=0.5)
    ax.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)
    # ============================
    plt.tight_layout()
    plt.show()
```
